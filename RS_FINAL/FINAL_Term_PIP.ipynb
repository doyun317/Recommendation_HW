## PIP 

```python
import numpy as np
import pandas as pd
```

```python
def Agreement(r1,r2):
    R_max = 1
    R_max = 0
    R_med = 0.5
    
    mask = np.full((1,100),0.5)
    
    k= ((r1>mask) & (r2<mask)) | ((r1<mask) & (r2>mask))

    return k
```

```python
def proximity(r1,r2):
    R_max = 1
    R_max = 0
    R_med = 0.5
    D = np.array([])
    TF = Agreement(r1,r2)
    
    for i in range(100):
        if TF[0,i]:
            D = np.append(D,np.abs(r1[i]-r2[i]))
        else :
            D = np.append(D,2*np.abs(r1[i]-r2[i]))

    return D
```

```python
def impact(r1,r2):
    R_max = 1
    R_max = 0
    R_med = 0.5
    I = np.array([])
    TF = Agreement(r1,r2)
    
    for i in range(100):
        if TF[0,i]:
            I = np.append(I,(np.abs(r1[i]-R_med)+1)*(np.abs(r2[i]-R_med)+1))
        else :
            I = np.append(I,1/((np.abs(r1[i]-R_med)+1)*(np.abs(r2[i]-R_med)+1)))

    
    return I
```

```python
def popularity(r1,r2,u_k):
    
    mask = u_k
    TF = ((r1>mask) & (r2>mask)) | ((r1<mask) & (r2<mask))
    P = np.array([])

    
    for i in range(100):
        if TF[i] :
            P = np.append(P,1+((r1[i]+r2[i])/2 - u_k[i])**2)
        else :
            P = np.append(P,1)
        
    return P
```

```python
def pip(r1,r2,u_k):
    return proximity(r1,r2) * impact(r1,r2) * popularity(r1,r2,u_k)
```

```python
def pip1(r1,r2,u_k):
    print(proximity(r1,r2))
    print(impact(r1,r2))
    print(popularity(r1,r2,u_k))
```
