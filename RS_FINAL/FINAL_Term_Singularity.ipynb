## SINGULARITY

```python
SP = data[data>=0.5].count()/data.count()
SN = data[data<0.5].count()/data.count()
```

```python
empty = np.zeros((2,100))

empty[0] = SP
empty[1] = SN

SP_SN = pd.DataFrame(empty.T,columns = ['SP','SN'])
SP_SN 
```

```python
datasize = 20 # 200 * 200 연산은 시간소모가 너무 커 20명에 대한 정보만 사용

SM_ = pd.DataFrame(np.zeros((datasize,datasize)))
spsn = SP_SN 

for i in data.index.values[:datasize]:
    for j in data.index.values[:datasize]:
        if(i==j):
            continue
        A_sum = 0
        B_sum = 0
        C_sum = 0
        _sum = 0
        
        A_size = 0
        B_size = 0
        C_size = 0
        
     ### A 연산
        _A = ((data.iloc[i,:]>=0.5) & (data.iloc[j,:]>=0.5)) 
        arr = [temp for temp , x in enumerate(_A) if x]
        
        if(_A.value_counts()[False] == 100) :
            continue
        else:
            A_size = _A.value_counts()[True]

            for k in arr:

                temp_sum = (1 - (data.iloc[i,k] - data.iloc[j,k])**2)*spsn["SP"][k]*spsn["SP"][k]

                if (np.isnan(temp_sum)):
                    A_size = A_size-1
                    continue
                else:
                    A_sum += temp_sum
        
    
     ### B 연산
        _B = ((data.iloc[i,:]<0.5) & (data.iloc[j,:]<0.5)) 
        arr = [temp for temp , x in enumerate(_B) if x]
        
        if(_B.value_counts()[False] == 100) :
            continue
        else:
            B_size = _B.value_counts()[True]

            for k in arr:

                temp_sum = (1 - (data.iloc[i,k] - data.iloc[j,k])**2)*spsn["SN"][k]*spsn["SN"][k]

                if (np.isnan(temp_sum)):
                    B_size = B_size-1
                    continue
                else:
                    B_sum += temp_sum
                
     ### C 연산
    
        _C = ((data.iloc[i,:]>=0.5) & (data.iloc[j,:]<0.5) | (data.iloc[i,:]<0.5) & (data.iloc[j,:]>=0.5)) 
        arr = [temp for temp , x in enumerate(_C) if x]
        

        
        if(_C.value_counts()[False] == 100) :
            continue
        else:
            C_size = _C.value_counts()[True]
  
            for k in arr:

                temp_sum = (1 - (data.iloc[i,k] - data.iloc[j,k])**2)*spsn["SP"][k]*spsn["SN"][k]

                if (np.isnan(temp_sum)):
                    C_size = C_size-1
                    continue
                else:
                    C_sum += temp_sum

                
                
        if(A_size>0):
            _sum += A_sum/A_size 

        if(B_size>0):
            _sum += B_sum/B_size 
            
        if(C_size>0):
            _sum += C_sum/C_size 
        
        SM_.iloc[i,j] = _sum * 1/3
```
